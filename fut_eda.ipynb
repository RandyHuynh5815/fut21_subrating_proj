```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8,8))
p = [("Attacker", (0,0)), ("Midfielder", (0,1)), ("Defender",(1,0)), ("Goalkeeper", (1,1))]
for pos, loc in p:
    labels, count = np.unique(df['WF'][df['Position'] == pos], return_counts=True)
    x,y = loc
    
    ax[x][y].pie(count, labels=labels, autopct=lambda x: '{:.2f}%'.format(x))
    ax[x][y].set_title(f"Weak Foot of {pos}s")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(nrows=5,ncols=1, figsize=(5,15))
for i in range(5):
    labels, poscount = np.unique(df["Position"][df['Skill'] == (i+1)], return_counts=True)
    
    if i == 0:
        ax[i].pie(poscount, labels=labels, autopct=lambda x: '{:.2f}%'.format(x), colors=['red'])
    else:
        ax[i].pie(poscount, labels=labels, autopct=lambda x: '{:.2f}%'.format(x))
    ax[i].set_title(f"Positions of {i+1} Starred Skill Moves")
plt.tight_layout()
plt.show()
```

# Subrating Correlation

Code from https://medium.com/@szabo.bibor/how-to-create-a-seaborn-correlation-heatmap-in-python-834c0686b88e

```python
df_field_corr = df[['PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY', 'WF', 'Skill']].corr()
sns.heatmap(df_field, vmin=-1, vmax=1, annot=True, cmap='BrBG')
```
